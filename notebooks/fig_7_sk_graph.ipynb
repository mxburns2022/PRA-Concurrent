```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import os
import sys
import plot_settings
import matplotlib as mpl
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif']= 'STIXGeneral'
mpl.rcParams['mathtext.fontset']= 'stix'
mpl.rcParams['font.size']= 28
import matplotlib.ticker as tk
```

```python
brim_df = pd.read_csv('../data/sk_full.csv').rename(columns={'blocks': 'Subsystems'})
brim_df['Model'] = brim_df['concurrent'].replace({True: 'Conc.', False: 'Serial'})
fig, axs = plt.subplots(ncols=2, figsize=(12,6), gridspec_kw=dict(wspace=0.02))
palette = [ "#404040",
            "#ef8a62",
            "#999999",
            "#ca0020"
            ]
# axs = [ax]
brim_df['Bandwidth'] = 1 / brim_df['tpart'] / 1e6
brim_df['error'] = brim_df['u'] - -0.7633
line_args = dict(
    y='error',
    x='tpart',
    linestyle='solid',
    # marker='X',
    markers=True,
    markersize=12,
    markeredgecolor='black',
    marker='X',
    linewidth=4,
    hue='Subsystems',
    err_style='bars',
    markevery=2,
    palette=palette
)
sns.lineplot(ax=axs[0],
             data=brim_df.query('nodes==2000 and (concurrent== True) and (Subsystems == 1 or Subsystems == 2 or Subsystems == 4 or Subsystems == 8)'), 
             **line_args)
sns.lineplot(ax=axs[1],
             data=brim_df.query('nodes==2000 and (concurrent== False or Subsystems==1) and (Subsystems == 1 or Subsystems == 2 or Subsystems == 4 or Subsystems == 8)'), 
             **line_args)
axs[1].legend(fontsize=22)
for ax, label in zip(axs, ['a', 'b', 'c', 'd']):
    ax.set_yscale('log')
    ax.set_xlabel('Sync Frequency ($\\tau^{-1}$, MHz)')
    ax.set_ylim(6e-3, 3e-1)
    ax.grid(which='minor', linestyle='--', alpha=0.5)
    ax.grid(which='major', linewidth=1)
    ax.annotate(text=f"({label})", xy=(0.45, -0.28), xycoords='axes fraction', fontsize=28)
print(axs[0].get_xticks())
for ax in axs[1:]:
    ax.set_ylabel('')
    ax.set_yticklabels(['' for i in ax.get_yticks()])
for ax in axs[0:-1]:
    ax.legend_ = None
axs[0].set_ylabel('$u - u^\\star$', fontsize=40)
axs[0].yaxis.set_label_coords(-0.1,0.45)
fig.suptitle('$N=2000$ SK Energy Error vs. $\\tau$')
axs[-1].legend(fontsize=24, title='Subsystems')
axs[0].annotate('Concurrent', xy=(0.14, .8), xycoords='subfigure fraction', weight='bold', fontsize=30)
axs[1].annotate('Serial', xy=(0.615, .8), xycoords='subfigure fraction', weight='bold', fontsize=30)
axs[0].set_xlabel("Epoch ($\\tau$)")
axs[1].set_xlabel("Epoch ($\\tau$)")
fig.savefig('../figures/fig_7_sk_comparison.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
